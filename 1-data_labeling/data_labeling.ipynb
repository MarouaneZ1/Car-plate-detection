# from XML to CSV

```python
import pandas as pd
import xml.etree.ElementTree as xet
from glob import glob
```

```python
path = glob('./images/*.xml')
path
```

```python
len(path)
```

```python
filename = path[0]
info = xet.parse(filename)
root = info.getroot()
member_object = root.find('object')
labels_info = member_object.find('bndbox')
ymax = int(labels_info.find('ymax').text)
ymin = int(labels_info.find('ymin').text)
xmax = int(labels_info.find('xmax').text)
xmin = int(labels_info.find('xmin').text)
```

```python
print(ymax,ymin,xmax,xmin)
```

```python
cars_infos = dict(filepath=[], xmin=[], xmax=[], ymin=[], ymax=[])
for filename in path:
    info = xet.parse(filename)
    root = info.getroot()
    member_object = root.find('object')
    labels_info = member_object.find('bndbox')
    ymax = int(labels_info.find('ymax').text)
    ymin = int(labels_info.find('ymin').text)
    xmax = int(labels_info.find('xmax').text)
    xmin = int(labels_info.find('xmin').text)
    cars_infos['filepath'].append(filename)
    cars_infos['xmin'].append(xmin)
    cars_infos['xmax'].append(xmax)
    cars_infos['ymin'].append(ymin)
    cars_infos['ymax'].append(ymax)
```

```python
df = pd.DataFrame(cars_infos)
df
```

```python
df.to_csv('CarsImagesLabeled.csv', index=False)
```
